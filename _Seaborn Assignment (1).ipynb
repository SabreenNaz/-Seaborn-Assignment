Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

#Answer


Seaborn is a powerful Python data visualization library based on Matplotlib. Here are five types of plots that  we create using Seaborn:


1) Scatter Plot with Regression Line:

* Seaborn Function: seaborn.regplot()
* Use: Visualize the relationship between two variables and show a regression line to represent the trend.

2) Pair Plot:

* Seaborn Function: seaborn.pairplot()
* Use: Explore relationships between variables in a dataset, especially useful for multivariate analysis.

3) Heatmap:

* Seaborn Function: seaborn.heatmap()
* Use: Visualize the correlation matrix or any 2D dataset to highlight patterns and relationships.

4) Box Plot:

* Seaborn Function: seaborn.boxplot()
* Use: Display the distribution of data across categories and highlight statistical summaries.

5) Violin Plot:

* Seaborn Function: seaborn.violinplot()
* Use: Similar to a box plot but provides a deeper insight into the distribution across different categories.

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

Note: pclass, age, and fare are columns in the titanic dataset.

```python
#Answer

import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Set the style
sns.set(style="whitegrid")

# Create subplots for two box plots
plt.figure(figsize=(12, 6))

# Box plot for 'age'
plt.subplot(1, 2, 1)
sns.boxplot(x="pclass", y="age", data=titanic_data)
plt.title("Box Plot of Age by Pclass")

# Box plot for 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(x="pclass", y="fare", data=titanic_data)
plt.title("Box Plot of Fare by Pclass")

# Show the plots
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
#Answer

import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Set the style
sns.set(style="whitegrid")

# Create a histogram with hue for the 'cut' column
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x="price", hue="cut")

# Set plot labels and title
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Histogram of Diamond Prices by Cut")

# Show the plot
plt.show()

```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
#Answer

import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix for the heatmap
flights_matrix = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Set the style
sns.set(style="whitegrid")

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(flights_matrix, annot=True, cmap="YlGnBu", fmt="d", linewidths=.5)

# Set plot labels and title
plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Number of Passengers by Month and Year")

# Show the plot
plt.show()
```
